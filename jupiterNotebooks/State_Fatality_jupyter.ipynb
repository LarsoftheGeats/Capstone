```python
def read_in_excel(year):
    data=pd.read_excel('Fatalities_by_state.xlsx',header=3,sheet_name=year)
    data.columns=['state',
            'rural_interstate',
            'urban_interstate',
            'freeway_expressway',
            'other_arterial',
            'minor_arterial',
            'collector_arterial',
            'local',
            'unknown',
            'total']
    data['year']=int(year)
#     data.set_index(['year','state'], inplace=True)
    return data
```
