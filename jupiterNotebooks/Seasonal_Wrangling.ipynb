```python
#revised funciton
def summary_in(month, year):
    #sheet_name is the second sheet.  header drops excel formating
    data=pd.read_excel(f'tvt{month}{year}.xls',sheet_name="Tables 1 & 2",header=7)
    #slicing the dataframe
    data=data.iloc[1:16,:]
    #dropping some formatting rows
    data.drop((8), axis=0,inplace=True)
    data.set_index('System',drop=True,inplace=True)
    #transpose
    data=data.transpose()
    data2=data.iloc[:,7:]
    data=data.iloc[:,:7]

    data=data.append(data2)
    year_int=int(year)-1
    year_str=str(year_int).rjust(2,'0')
    data.index=pd.period_range(start=f"jan 20{year_str}", end =f'dec 20{year}', freq = 'm')
    return data
```
